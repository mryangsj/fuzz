## Import Necessary Libraries


```python
import h5py
import numpy as np
import matplotlib.pyplot as plt
from matplotlib import cm
from matplotlib.ticker import LinearLocator
%matplotlib widget
```

## Read H5 File


```python
z_path = "../Z.hdf5"
with h5py.File(z_path, "r") as file:
    z_matlab = file["/Z"][:].T

fs_dict = {
    44.1e3 : 0,
    48e3   : 1,
    88.2e3 : 2,
    96e3   : 3,
    176.4e3: 4,
    192e3  : 5
}

z_44_1 = z_matlab[fs_dict[44.1e3]]
z_48 = z_matlab[fs_dict[48e3]]
z_88_2 = z_matlab[fs_dict[88.2e3]]
z_96 = z_matlab[fs_dict[96e3]]
z_176_4 = z_matlab[fs_dict[176.4e3]]
z_192 = z_matlab[fs_dict[192e3]]

z_root_44_1 = z_44_1[:, :, 0:2, 0:2]
z_root_48 = z_48[:, :, 0:2, 0:2]
z_root_88_2 = z_88_2[:, :, 0:2, 0:2]
z_root_96 = z_96[:, :, 0:2, 0:2]
z_root_176_4 = z_176_4[:, :, 0:2, 0:2]
z_root_192 = z_192[:, :, 0:2, 0:2]
```

```python
# xy
R_lin = np.arange(0, 101, 1)
R_vol = np.arange(0, 101, 1)
X, Y = np.meshgrid(R_lin, R_vol)

plt.close('all')
fig, axs = plt.subplots(2, 3, figsize=(12, 7), subplot_kw={'projection': '3d'})

axs[0, 0].plot_surface(X, Y, z_root_44_1[:, :, 0, 0], cmap=cm.coolwarm)
axs[0, 0].set_xlabel(r'$R_{lin}$ [%]')
axs[0, 0].set_ylabel(r'$R_{vol}$ [%]')
axs[0, 0].set_title(r'$Z_{11}$')

axs[0, 1].plot_surface(X, Y, z_root_44_1[:, :, 0, 1], cmap=cm.coolwarm)
axs[0, 1].set_xlabel(r'$R_{lin}$ [%]')
axs[0, 1].set_ylabel(r'$R_{vol}$ [%]')
axs[0, 1].set_title(r'$Z_{12}$, $Z_{21}$')

axs[0, 2].plot_surface(X, Y, z_root_44_1[:, :, 1, 1], cmap=cm.coolwarm)
axs[0, 2].set_xlabel(r'$R_{lin}$ [%]')
axs[0, 2].set_ylabel(r'$R_{vol}$ [%]')
axs[0, 2].set_title(r'$Z_{22}$')

axs[1, 0].plot_surface(X, Y, z_root_192[:, :, 0, 0], cmap=cm.coolwarm)
axs[1, 0].set_xlabel(r'$R_{lin}$ [%]')
axs[1, 0].set_ylabel(r'$R_{vol}$ [%]')
axs[1, 0].set_title(r'$Z_{11}$')

axs[1, 1].plot_surface(X, Y, z_root_192[:, :, 0, 1], cmap=cm.coolwarm)
axs[1, 1].set_xlabel(r'$R_{lin}$ [%]')
axs[1, 1].set_ylabel(r'$R_{vol}$ [%]')
axs[1, 1].set_title(r'$Z_{12}$, $Z_{21}$')

axs[1, 2].plot_surface(X, Y, z_root_192[:, :, 1, 1], cmap=cm.coolwarm)
axs[1, 2].set_xlabel(r'$R_{lin}$ [%]')
axs[1, 2].set_ylabel(r'$R_{vol}$ [%]')
axs[1, 2].set_title(r'$Z_{22}$')

plt.tight_layout()
plt.show()
```

```python
# xy
R_lin = np.arange(0, 101, 1)
R_vol = np.arange(0, 101, 1)
X, Y = np.meshgrid(R_lin, R_vol)

plt.close('all')
fig, axs = plt.subplots(2, 3, figsize=(12, 7), subplot_kw={'projection': '3d'})

axs[0, 0].plot_surface(X, Y, z_root_44_1[:, :, 0, 0] + z_root_44_1[:, :, 0, 1], cmap=cm.coolwarm)
axs[0, 0].set_xlabel(r'$R_{lin}$ [%]')
axs[0, 0].set_ylabel(r'$R_{vol}$ [%]')
axs[0, 0].set_title(r'$R_{1}$')

axs[0, 1].plot_surface(X, Y, z_root_44_1[:, :, 1, 1] + z_root_44_1[:, :, 0, 1], cmap=cm.coolwarm)
axs[0, 1].set_xlabel(r'$R_{lin}$ [%]')
axs[0, 1].set_ylabel(r'$R_{vol}$ [%]')
axs[0, 1].set_title(r'$R_{2}$')

axs[0, 2].plot_surface(X, Y, -z_root_44_1[:, :, 0, 1], cmap=cm.coolwarm)
axs[0, 2].set_xlabel(r'$R_{lin}$ [%]')
axs[0, 2].set_ylabel(r'$R_{vol}$ [%]')
axs[0, 2].set_title(r'$R_{3}$')

axs[1, 0].plot_surface(X, Y, z_root_192[:, :, 0, 0] + z_root_192[:, :, 0, 1], cmap=cm.coolwarm)
axs[1, 0].set_xlabel(r'$R_{lin}$ [%]')
axs[1, 0].set_ylabel(r'$R_{vol}$ [%]')
axs[1, 0].set_title(r'$R_{1}$')

axs[1, 1].plot_surface(X, Y, z_root_192[:, :, 1, 1] + z_root_192[:, :, 0, 1], cmap=cm.coolwarm)
axs[1, 1].set_xlabel(r'$R_{lin}$ [%]')
axs[1, 1].set_ylabel(r'$R_{vol}$ [%]')
axs[1, 1].set_title(r'$R_{2}$')

axs[1, 2].plot_surface(X, Y, -z_root_192[:, :, 0, 1], cmap=cm.coolwarm)
axs[1, 2].set_xlabel(r'$R_{lin}$ [%]')
axs[1, 2].set_ylabel(r'$R_{vol}$ [%]')
axs[1, 2].set_title(r'$R_{3}$')

plt.tight_layout()
plt.show()
```

## Report the Range of Impedance


```python
z11_min = None
z11_max = None
z12_min = None
z12_max = None
z21_min = None
z21_max = None
z22_min = None
z22_max = None

for fs in fs_dict:
    z = z_matlab[fs_dict[fs]]
    z11 = z[:, :, 0, 0]
    z12 = z[:, :, 0, 1]
    z22 = z[:, :, 1, 1]

    r1 = z11 + z12
    r2 = z22 + z12
    r3 = -z12

    z11_max_index = np.unravel_index(np.argmax(z11), z11.shape)
    z12_max_index = np.unravel_index(np.argmax(z12), z12.shape)
    z22_max_index = np.unravel_index(np.argmax(z22), z22.shape)

    z11_min_index = np.unravel_index(np.argmin(z11), z11.shape)
    z12_min_index = np.unravel_index(np.argmin(z12), z12.shape)
    z22_min_index = np.unravel_index(np.argmin(z22), z22.shape)

    r1_max_index = np.unravel_index(np.argmax(r1), r1.shape)
    r2_max_index = np.unravel_index(np.argmax(r2), r2.shape)
    r3_max_index = np.unravel_index(np.argmax(r3), r3.shape)

    r1_min_index = np.unravel_index(np.argmin(r1), r1.shape)
    r2_min_index = np.unravel_index(np.argmin(r2), r2.shape)
    r3_min_index = np.unravel_index(np.argmin(r3), r3.shape)

    if z11_max is None or z11[z11_max_index] > z11_max:
        z11_max = z11[z11_max_index]
    if z11_min is None or z11[z11_min_index] < z11_min:
        z11_min = z11[z11_min_index]
    if z12_max is None or z12[z12_max_index] > z12_max:
        z12_max = z12[z12_max_index]
    if z12_min is None or z12[z12_min_index] < z12_min:
        z12_min = z12[z12_min_index]
    if z21_max is None or z12[z12_max_index] > z21_max:
        z21_max = z12[z12_max_index]
    if z21_min is None or z12[z12_min_index] < z21_min:
        z21_min = z12[z12_min_index]
    if z22_max is None or z22[z22_max_index] > z22_max:
        z22_max = z22[z22_max_index]
    if z22_min is None or z22[z22_min_index] < z22_min:
        z22_min = z22[z22_min_index]

    print("=======================================")
    print(f"fs: {fs}")
    print(f"z11: {z11[z11_min_index]:.4f} ~ {z11[z11_max_index]:.4f}")
    print(f"z22: {z22[z22_min_index]:.4f} ~ {z22[z22_max_index]:.4f}")
    print(f"z12, z21: {z12[z12_min_index]:.4f} ~ {z12[z12_max_index]:.4f}")

    # print("\n")
    # print(f"r1: {r1[r1_min_index]:.4f} ~ {r1[r1_max_index]:.4f}")
    # print(f"r2: {r2[r2_min_index]:.4f} ~ {r2[r2_max_index]:.4f}")
    # print(f"r3: {r3[r3_min_index]:.4f} ~ {r3[r3_max_index]:.4f}")
    # print("\n")

print("=======================================")
print(f"z11: {z11_min:.4f} ~ {z11_max:.4f}")
print(f"z22: {z22_min:.4f} ~ {z22_max:.4f}")
print(f"z12, z21: {z12_min:.4f} ~ {z12_max:.4f}")
```
